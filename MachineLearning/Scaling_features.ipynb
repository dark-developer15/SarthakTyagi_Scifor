### Scaling Features

Using Sk-learn library

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
```

```python
# importing the data
arr = pd.read_csv("../content/sample_data/mnist_train_small.csv").iloc[:100, :12]

# Reading 100 rows and 12 columns beacuse this dataset is too big for this task
print(arr)
```

```python
# creating the object of feature
feature = MinMaxScaler()

# using fit_transform to scale the feature
scaled_feature = feature.fit_transform(arr)

print(scaled_feature)
```

```python
# Self-Learning
# Using the Standard Deviation for the same data...

feature = StandardScaler()

scaled_feature = feature.fit_transform(arr)

print(scaled_feature)
```

```python
import matplotlib.pyplot as plt

# for each column creating a plot
for col in arr.columns:
    plt.plot(arr.index, arr[col], label=col)

plt.show()
```

```python
# visualising data after scaling using Standard scaler technique
plt.plot(scaled_feature)

plt.show()
```
