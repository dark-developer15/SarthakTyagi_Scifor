```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/sarth/OneDrive/Desktop/Jupyter/Datasets/BikeTripDataSet/Cleaned_Data.csv", index_col=0)
df.info()
```

```python
# removing duplicate values from the station_start_name column
df = df.drop_duplicates(subset=df.columns[4])
print(df.drop_duplicates(subset=df.columns[6]).info())
print(df.drop_duplicates(subset=df.columns[4]).info())
```

```python
# Creating a chart to visualize unique values after removing duplicates
length = []
columns = []
for i in df.columns:
    length.append(len(df.drop_duplicates(subset=i)))
    columns.append(i)

plt.figure(figsize=(15, 10))
# plt.bar(columns, length, color='cyan')
plt.plot(columns, length, 'o-', color='red', label='Unique Row Lengths')
for i, txt in enumerate(length):
    plt.annotate(txt, (columns[i], length[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel("Columns")
plt.ylabel("Length of Unique Rows")
plt.title("Distribution of Unique Row Lengths after Removing Duplicate Values from Each Column")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
